```
import matplotlib.pyplot as plt
import math
```

Equação logística com x0 = 0,64 e taxa de crescimento populacional (a) fixa em 1,3.

```
t = 1
a = 1.3
x0 = 0.64
plt.axis([0, 20, 0, 0.8])
for t in range(101):
  x = a*x0*(1-x0)
  plt.scatter(t, x0)
  x0 = x
```

```
t = 1
a = 1.3
x0 = 0.64
for t in range(101):
  x = a*x0*(1-x0)
  plt.scatter(t, math.log(x0-(1-1/a)))
  x0 = x
```

Agora com a taxa de crescimento populacional (a) fixa em 0,8.

```
t = 1
a = 0.8
x0 = 0.64
plt.axis([0, 20, 0, 0.8])
for t in range(101):
  x = a*x0*(1-x0)
  plt.scatter(t, x0)
  x0 = x
```

```
t = 1
a = 0.8
x0 = 0.64
for t in range(101):
  x = a*x0*(1-x0)
  plt.scatter(t, math.log(x0))
  x0 = x
```

Agora com a taxa de crescimento populacional (a) crescendo de 0,5 em 0,5.

```
t = 1
a = 1.1
x0 = 0.64
for t in range(101):
  x = a*x0*(1-x0)
  plt.scatter(t, x0)
  x0 = x
  a = a + 0.5
```
