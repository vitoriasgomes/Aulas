#Gráficos de função

##Gráfico de sen(x) e sen(x)+0.5

```
import matplotlib.pyplot as plt
from numpy import sin, pi
import numpy as np
```

```
x = np.linspace(-pi,pi,10001)
y = np.sin(x)
```

```
plt.plot(x,y,"b-")
plt.plot(x,y+0.5, "g-")
plt.savefig("seno")
```
